```python
plt.figure(figsize=(6,4))
sns.histplot(train["outcome"], bins=50, kde=True)
plt.title("Distribution of outcome")
plt.tight_layout()
plt.savefig("../report/figures/target_dist.png")
plt.show()
```

# Is it skewed?
# Heavy-tailed?
# Any extreme outliers?

```python
cat_cols = ["cut", "color", "clarity"]

for col in cat_cols:
    display(train[col].value_counts())
```

# How many levels?
# Any very rare categories? 

```python
num_cols = train.drop(columns=cat_cols + ["outcome"])
corr = num_cols.corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Numeric feature correlation heatmap")
plt.tight_layout()
plt.savefig("../report/figures/corr_heatmap.png")
plt.show()
```
